### Here is the new dataframe with the average temp of all cities

```python
Average = pd.concat([Mumbai['date_time'], All_temp ], axis=1, keys=['date_time', 'Avg_temp'])
```

```python
# Printing the first three rows to verify the dataframe. 
# Average.head(3)
```

```python
# Creating a new column with the rounded Average temp
Average['Avg_temp_rounded'] = round(Average['Avg_temp'])
```

### Plotting a line graph for the Average temp of all 50 cities

```python
plt.figure(figsize=(16,6))
plt.plot(Average.date_time, Average.Avg_temp_rounded, linestyle='--', marker='o')
plt.xlabel("5 days average temp of all cities for every 3 hour window")
plt.xticks(rotation = 90)
plt.ylabel("Temperature (C)")
plt.title("Average temp of top 50 cities based on population")
plt.show()
```

    The Coldest night for all the cities is on 10th of january and the hottest day will be on 12th of january in a timeframe of 09 january to 14 january.
